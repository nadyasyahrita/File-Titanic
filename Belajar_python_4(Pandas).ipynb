###**Data Frame**

```
import numpy as np
import pandas as pd
```

```
np.random.seed(100) #generate bilangan random
```

```
dataku=pd.DataFrame(np.random.randn(3,5),['A','B','C'],['satu','dua','tiga','empat','lima']) #urutan pertama adalah baris lalu kolom
```

```
dataku
```

```
#memilih sebuah kolom
dataku['dua'] 
```

```
type(dataku['dua'])
```

```
#menambahkkan kolom baru
dataku['baru']=dataku['empat']+dataku['lima']
dataku
```

```
#menghapus sebuah kolom dengan menggunakan drop
dataku.drop('tiga',axis=1)
```

```
#lalu di assign agar tersimpan yg kolom-3 sudah dihilagkan
dataku=dataku.drop('tiga',axis=1)
dataku
```

```
#cara lain untuk menghapus kolom baru
dataku.drop('baru',axis=1,inplace=True)
dataku
```

```
#cara untuk menghapus baris dengan inplace
dataku.drop('B',axis=0,inplace=True)
dataku
```

```
dataku.shape #ada dua baris dan 4 kolom
```
